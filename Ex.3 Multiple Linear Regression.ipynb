```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import metrics
boston = datasets.load_boston()
print(boston.data.shape)
print(boston.target.shape)
print(boston.feature_names)
df = pd.DataFrame(boston.data)
print(df.head())
df.columns = boston.feature_names
print(df.head(2))
print(boston.target.shape)
df[&#39;PRICE&#39;] = boston.target
print(df.head(2))
df.describe()
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target,
test_size=0.10, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
model = LinearRegression()
model.fit(X_train, y_train)
print(model.coef_)
print(&#39;Intercept: %.3f&#39; % model.intercept_)
y_pred = model.predict(X_test)
print(y_pred[:5]) #first 5 predictions
print(y_test[:5]) #first 5 target values
plt.scatter(y_test, y_pred)
plt.xlabel(&#39;Price&#39;)
plt.ylabel(&#39;Predicted Price&#39;)
plt.title(&#39;Boston House Price vs Predicted Price&#39;)
print(&#39;R-squared value on training dataset: %0.3f&#39; %model.score(X_train, y_train))
print(&#39;R-squared value on training dataset: %0.3f&#39; %model.score(X_test, y_test))
print(&#39;Mean Absolute Error: %0.3f&#39; %metrics.mean_absolute_error(y_test, y_pred))
print(&#39;Mean Squared Error: %0.3f&#39; %metrics.mean_squared_error(y_test, y_pred))
print(&#39;Root Mean Squared Error: %0.3f&#39; %np.sqrt(metrics.mean_squared_error(y_test,
y_pred)))
mean = np.mean(y_test)
print(&#39;10% of Mean value of target: {:0.3f}&#39;.format(np.mean(y_test)*0.1))
```
